```python
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv("https://raw.githubusercontent.com/FairozAhmadSheikh/Datasets_CSV/refs/heads/main/Placement.csv")
df = df.drop(columns=['Student_ID'], axis=1)
```

```python
print(df.head())
```

```python
sns.scatterplot(x=df['CGPA'], y=df['IQ'],hue=df['Placement'])
plt.xlabel("CGPA")
plt.ylabel("IQ")
plt.title("Scatterplot of CGPA vs IQ")
plt.show()
```

```python
from sklearn.linear_model import Perceptron
from sklearn.preprocessing import StandardScaler
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt


X = df.iloc[:, 0:2].values  
y = df['Placement'].values 

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


p = Perceptron()
p.fit(X_scaled, y)

p.coef_ 
p.intercept_  
# Plot 

plot_decision_regions(X_scaled, y, clf=p, legend=2)
plt.xlabel("CGPA (scaled)")
plt.ylabel("IQ (scaled)")
plt.title("Perceptron Decision Boundary after Scaling")
plt.show()
```
